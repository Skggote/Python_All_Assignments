
# **Admin Code**




```python
class FoodItem:
    next_food_id = 1

    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = FoodItem.next_food_id
        FoodItem.next_food_id += 1
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

food_items = []

def add_food_item():
    name = input("Enter food item name: ")
    quantity = input("Enter quantity: ")
    price = float(input("Enter price: "))
    discount = float(input("Enter discount: "))
    stock = int(input("Enter stock: "))

    food_item = FoodItem(name, quantity, price, discount, stock)
    food_items.append(food_item)
    print(f"Food item added successfully with FoodID: {food_item.food_id}")

def edit_food_item(food_id):
    for food_item in food_items:
        if food_item.food_id == food_id:
            print(f"Editing Food Item: {food_item.name}")
            print("1. Edit Name")
            print("2. Edit Price")
            print("3. Edit Stock")
            print("4. Back to Admin Menu")
            choice = int(input("Enter your choice: "))

            if choice == 1:
                new_name = input("Enter the new name: ")
                food_item.name = new_name
                print("Name updated successfully!")
            elif choice == 2:
                new_price = float(input("Enter the new price: "))
                food_item.price = new_price
                print("Price updated successfully!")
            elif choice == 3:
                new_stock = int(input("Enter the new stock: "))
                food_item.stock = new_stock
                print("Stock updated successfully!")

                if new_stock == 0:
                    print("This item is now out of stock.")
            elif choice == 4:
                return
            else:
                print("Invalid choice")

def remove_food_item(food_id):
    for food_item in food_items:
        if food_item.food_id == food_id:
            food_items.remove(food_item)
            print(f"Food item with FoodID {food_id} removed successfully.")
            return
    print(f"Food item with FoodID {food_id} not found.")

def admin_menu():
    while True:
        print("\nAdmin Menu")
        print("1. Add new food item")
        print("2. Edit food item")
        print("3. View all food items")
        print("4. Remove food item")
        print("5. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            add_food_item()
        elif choice == 2:
            food_id = int(input("Enter FoodID to edit: "))
            edit_food_item(food_id)
        elif choice == 3:
            view_all_food_items()
        elif choice == 4:
            food_id = int(input("Enter FoodID to remove: "))
            remove_food_item(food_id)
        elif choice == 5:
            break
        else:
            print("Invalid choice")

def view_all_food_items():
    for food_item in food_items:
        print(f"FoodID: {food_item.food_id}, Name: {food_item.name}, Price: {food_item.price}, Stock: {food_item.stock}")

if __name__ == "__main__":
    admin_menu()
```

# **User Code:**

```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password

class Order:
    def __init__(self, user):
        self.user = user
        self.items = []
        self.total_amount = 0

    def add_item(self, food_item):
        self.items.append(food_item)
        self.total_amount += food_item.price

def place_order(user):
    print("Available Food Items:")
    for idx, item in enumerate(food_items, start=1):
        print(f"{idx}. {item.name} {item.quantity} [{item.price}]")

    selected_indices = list(map(int, input("Select food items: ").split(',')))
    selected_items = [food_items[idx - 1] for idx in selected_indices]

    order = Order(user)
    for item in selected_items:
        order.add_item(item)

    return order

def user_menu(user):
    while True:
        print("\nUser Menu")
        print("1. Place New Order")
        print("2. Order History")
        print("3. Update Profile")
        print("4. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            order = place_order(user)
            print("\nOrder placed successfully!")
            print("Items in the order:")
            for item in order.items:
                print(item.name)
                print(item.quantity)
            print("Total Amount:", order.total_amount)
        elif choice == 2:
            pass
        elif choice == 3:
            pass
        elif choice == 4:
            break
        else:
            print("Invalid choice")

if __name__ == "__main__":
    full_name = input("Enter Full Name: ")
    phone_number = input("Enter Phone Number: ")
    email = input("Enter Email: ")
    address = input("Enter Address: ")
    password = input("Enter Password: ")

    user = User(full_name, phone_number, email, address, password)
    user_menu(user)
```
